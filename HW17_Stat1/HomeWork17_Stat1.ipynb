```python
horse_8c = horse_8c[horse_8c.CapillaryRefillTime.between(0, 2, inclusive=True)]
```

Есть информация, что мксимальный пульс при тяжелой активности или при заболеваниях 120-130 (в некоторых источниках до 150).

```python
q1 = horse_8c.Pulse.quantile(0.25)
q3 = horse_8c.Pulse.quantile(0.75)
iqr = q3 - q1
upper_bound = q3 + (1.5 * iqr)
print(upper_bound)
remove_outliers = horse_8c[horse_8c.Pulse.between(0, upper_bound, inclusive=True)].sort_values('Pulse')
remove_outliers
```

```python
print(remove_outliers.Pulse.mean())
print(remove_outliers.Pulse.median())
```

Среднее значение изменилось незначительно (менее 3%), а медиана осталась на прежнем уровне. Полагаю фильтрация в данном случае допустима.

```python
q1 = remove_outliers.RespiratoryRate.quantile(0.25)
q3 = remove_outliers.RespiratoryRate.quantile(0.75)
iqr = q3 - q1
upper_bound = q3 + (1.5 * iqr)
print(upper_bound)
remove_outliers2 = remove_outliers[remove_outliers.RespiratoryRate.between(0, upper_bound, inclusive=True)] \
    .sort_values('RespiratoryRate')
remove_outliers2
```

```python
print('Средняя частота дыхания (ЧД) до фильтрации выбросов ЧД:', remove_outliers.RespiratoryRate.mean())
print('Медиана частоты дыхания до фильтрации выбросов ЧД:', remove_outliers.RespiratoryRate.median())
print('Средняя частота дыхания после фильтрации выбросов ЧД:', remove_outliers2.RespiratoryRate.mean())
print('Медиана частоты дыхания после фильтрации выбросов ЧД:', remove_outliers2.RespiratoryRate.median())
```

```python
print('Средний пульс до фильтрации выбросов ЧД:',remove_outliers.Pulse.mean())
print('Медиана пульса до фильтрации выбросов ЧД:', remove_outliers.Pulse.median())
print('Средний пульс после фильтрации выбросов ЧД:', remove_outliers2.Pulse.mean())
print('Медиана пульса после фильтрации выбросов ЧД:', remove_outliers2.Pulse.median())
```

#### Температура конечностей

```python
remove_outliers.loc[(remove_outliers.TemperatureOfExtremities.isnull())]
```

В файле с описанием сказано, что горячие кончнености должны коррелировать с высокой ректальной температурой, поэтому пропуски с температурой кончностей будем зполнять в зависимости от температуры. Нормальный диапазон:
1. Для взрослых 37.5-38.5
2. Для жеребят 37.5-39.0
Источник информации: https://www.allvet.ru/guide/guide7/

То есть, если температура выше диапазона, выстявляем WARM (2); если в диапазоне, то NORMAL (1)


```python
remove_outliers.loc[(remove_outliers.TemperatureOfExtremities.isnull()) & (remove_outliers.Age == 1)
                   & (remove_outliers.RectalTemperature >= 38.5), 'TemperatureOfExtremities'] = 2
remove_outliers.loc[(remove_outliers.TemperatureOfExtremities.isnull()) & (remove_outliers.Age == 1)
                   & (remove_outliers.RectalTemperature < 38.5), 'TemperatureOfExtremities'] = 1
remove_outliers.loc[(remove_outliers.TemperatureOfExtremities.isnull()) & (remove_outliers.Age == 2)
                   & (remove_outliers.RectalTemperature >= 39), 'TemperatureOfExtremities'] = 2
remove_outliers.loc[(remove_outliers.TemperatureOfExtremities.isnull()) & (remove_outliers.Age == 2)
                   & (remove_outliers.RectalTemperature < 39), 'TemperatureOfExtremities'] = 1
remove_outliers.describe()
```
